## Popularity Based Recommender System

```python
# Merge books dataset with ratings
books_with_ratings = ratings.merge(books, on="ISBN")
```

```python
books_with_ratings
```

```python
# Prepare df for books with the count of ratings
books_ratings_count = books_with_ratings.groupby("Book-Title").count()['Book-Rating'].reset_index()
books_ratings_count.rename(columns={"Book-Rating": "Rating-Count"}, inplace=True)
books_ratings_count
```

```python
# Prepare df for books with average ratings
books_avg_ratings = books_with_ratings.groupby("Book-Title").mean()['Book-Rating'].reset_index()
books_avg_ratings.rename(columns={"Book-Rating": "Average-Rating"}, inplace=True)
books_avg_ratings
```

```python
# Prepare df for books with average ratings & count of ratings
books_avg_ratings_count = books_avg_ratings.merge(books_ratings_count, on="Book-Title")
books_avg_ratings_count
```

```python
# Prepare df for popular books which have highest average ratings with a minimum of 250 ratings count
popular_50_books = books_avg_ratings_count[books_avg_ratings_count["Rating-Count"] >= 250].sort_values("Average-Rating", ascending=False).head(50)
popular_books = books_avg_ratings_count[books_avg_ratings_count["Rating-Count"] >= 250].sort_values("Average-Rating", ascending=False)
popular_books
```

```python
popular_50_books
```

```python
popular_books_info = popular_books.merge(books, on="Book-Title")
popular_50_books_info = popular_50_books.merge(books, on="Book-Title")
```

```python
popular_50_books_info
```

```python
# Popular 50 books information
popular_50_unique_books_info = popular_50_books_info.drop_duplicates("Book-Title")
popular_50_unique_books_extracted_info = popular_50_books_info.drop_duplicates("Book-Title")[["Book-Title", "Book-Author", "Image-URL-L", "Average-Rating", "Rating-Count"]]
```

```python
popular_50_unique_books_info
```

```python
import pickle

popular_50_unique_books_extracted_info = popular_50_books_info.drop_duplicates("Book-Title")[["Book-Title", "Book-Author", "Image-URL-L", "Average-Rating", "Rating-Count"]]
pickle.dump(popular_50_unique_books_extracted_info, open("popular_50_unique_books_extracted_info.pkl", "wb"))
popular_50_unique_books_extracted_info
```

## Collaborative Filtering Based Recommender System

```python
books_with_ratings
```

```python
x = books_with_ratings.groupby("User-ID").count()["Book-Rating"] > 200
exp_users = x[x].index
exp_users
```

```python
filtered_exp_users = books_with_ratings[books_with_ratings["User-ID"].isin(exp_users)]
filtered_exp_users
```

```python
y = filtered_exp_users.groupby("Book-Title").count()["Book-Rating"] >= 50
famous_books = y[y].index
```

```python
filtered_exp_users_with_famous_books = filtered_exp_users[filtered_exp_users["Book-Title"].isin(famous_books)]
filtered_exp_users_with_famous_books
```

```python
filtered_exp_users_with_famous_books.duplicated().sum()
```

```python
filtered_exp_users_with_famous_books = filtered_exp_users_with_famous_books.drop_duplicates()
```

```python
final_df = filtered_exp_users_with_famous_books.pivot_table(index="Book-Title", columns="User-ID", values="Book-Rating")
```

```python
final_df.fillna(0, inplace=True)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(final_df)
similarity_score
```

```python
similarity_score.shape
```

```python
similarity_score[0]
```

```python
similarity_score[1]
```

```python
def recommend_books(book_name):
    # index fetch
    index = np.where(final_df.index == book_name)[0][0]
    similer_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    
    for i in similer_items:
        print(final_df.index[i[0]])
```

```python
def recommend_books_info(book_name):
    # index fetch
    index = np.where(final_df.index == book_name)[0][0]
    similer_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]
    
    data = []
    for i in similer_items:
        item = []
        temp_df = books[books["Book-Title"] == final_df.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Title"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Book-Author"].values))
        item.extend(list(temp_df.drop_duplicates("Book-Title")["Image-URL-M"].values))
        data.append(item)
        
    return data
```
